```python
data = {
    "intents": [
        {
            "tag": "greetings",
            "patterns": [
                "Hi",
                "Hey",
                "Is anyone there?",
                "Hello",
                "Hay",
                "What's up?"
            ],
            "responses": [
                "Hello, Planet Bot at your service!",
                "Hi there! How can I assist you today?"
            ]
        },
        {
            "tag": "goodbye",
            "patterns": [
                "Bye",
                "See you later",
                "Goodbye",
                "Catch you later"
            ],
            "responses": [
                "See you later, keep exploring the universe!",
                "Goodbye! Come back soon!"
            ]
        },
        {
            "tag": "planets",
            "patterns": [
                "Can you tell me about planets?",
                "What defines a planet?",
                "Explain planets",
                "Criteria for a planet?"
            ],
            "responses": [
                "A planet is a celestial body that meets specific criteria established by the International Astronomical Union (IAU) in 2006.\n"
                "According to the IAU, a planet must:\n"
                "1. Orbit a Star (or Stellar Remnant): A planet must orbit a star, like the Sun, or the remnants of one (e.g., a white dwarf or neutron star).\n"
                "2. Have Sufficient Mass: A planet must be massive enough for its gravity to shape it into a nearly round (spherical) form.\n"
                "3. Clear Its Orbital Path: A planet must have cleared its orbit of other debris, meaning it dominates its region of space.\n\n"
                "Types of Planets:\n"
                "- Terrestrial Planets: Rocky planets like Earth, Mars, Venus, and Mercury.\n"
                "- Gas Giants: Large planets made mostly of gases like Jupiter and Saturn.\n"
                "- Ice Giants: Planets like Uranus and Neptune, which have icy components.\n"
                "- Dwarf Planets: Bodies like Pluto that don't meet all criteria (e.g., they haven’t cleared their orbit)."
            ]
        },
        {
            "tag": "which",
            "patterns": [
                "Which are the planets in our solar system?",
                "Solar System",
                "List the planets",
                "How many planets are there?"
            ],
            "responses": [
                "Our Solar System includes eight known planets:\n"
                "- Mercury\n"
                "- Venus\n"
                "- Earth\n"
                "- Mars\n"
                "- Jupiter\n"
                "- Saturn\n"
                "- Uranus\n"
                "- Neptune"
            ]
        },
        {
            "tag": "math",
            "patterns": [
                "What is the diameters?",
                "Diameter?",
                "Can you tell me the size of planets?",
                "Planet diameters?"
            ],
            "responses": [
                "**Planetary Diameters:**\n"
                "- Mercury: 4,880 km (Smallest planet in the solar system)\n"
                "- Mars: 6,779 km (Known as the Red Planet)\n"
                "- Venus: 12,104 km (Similar in size to Earth, often called Earth's twin)\n"
                "- Earth: 12,742 km (Largest of the terrestrial planets)\n"
                "- Neptune: 49,244 km (Smallest of the gas/ice giants)\n"
                "- Uranus: 50,724 km (Slightly larger than Neptune)\n"
                "- Saturn: 120,536 km (Known for its stunning ring system)\n"
                "- Jupiter: 139,820 km (Largest planet in the solar system)"
            ]
        },
        {
            "tag": "stars",
            "patterns": [
                "Tell me about the stars",
                "What are stars",
                "Star?",
                "Explain stars",
                "What makes a star shine?"
            ],
            "responses": [
                "A star is a massive, luminous ball of gas (mostly hydrogen and helium) held together by its own gravity and powered by nuclear fusion in its core.\n"
                "Stars are the building blocks of galaxies and are responsible for producing most of the elements we find in the universe.\n"
                "They shine because nuclear fusion in their cores produces light and heat."
            ]
        },
        {
            "tag": "moons",
            "patterns": [
                "What are moons?",
                "Tell me about Earth's Moon",
                "Do planets have moons?",
                "Which planet has the most moons?"
            ],
            "responses": [
                "Moons are natural satellites that orbit planets. They vary in size, shape, and composition.\n"
                "**Fun Facts:**\n"
                "- Earth's Moon is the fifth-largest moon in the solar system.\n"
                "- Jupiter has the most moons, with 95 confirmed moons as of 2025!\n"
                "- Some moons, like Saturn’s Titan, have atmospheres."
            ]
        },
        {
            "tag": "space",
            "patterns": [
                "What is space?",
                "Tell me about outer space",
                "Define space",
                "What exists in space?"
            ],
            "responses": [
                "Space, or outer space, is the vast, seemingly infinite expanse beyond Earth’s atmosphere.\n"
                "It contains planets, stars, galaxies, black holes, nebulae, and more. It’s mostly a vacuum with some particles, radiation, and magnetic fields."
            ]
        },
        {
            "tag": "galaxies",
            "patterns": [
                "What is a galaxy?",
                "Tell me about galaxies",
                "What galaxy do we live in?",
                "How many galaxies are there?"
            ],
            "responses": [
                "A galaxy is a massive system of stars, gas, dust, and dark matter bound together by gravity.\n"
                "We live in the Milky Way Galaxy, which contains about 100-400 billion stars!\n"
                "There are estimated to be over 2 trillion galaxies in the observable universe."
            ]
        },
        {
            "tag": "blackholes",
            "patterns": [
                "What is a black hole?",
                "Explain black holes",
                "Can you tell me about black holes?",
                "What happens inside a black hole?"
            ],
            "responses": [
                "A black hole is a region in space where gravity is so strong that nothing, not even light, can escape.\n"
                "Black holes form when massive stars collapse at the end of their lifecycle.\n"
                "**Fun Fact:** The boundary of a black hole is called the event horizon. Beyond this, escape is impossible!"
            ]
        },
        {
            "tag": "help",
            "patterns": [
                "Could you help me?",
                "Give me a hand please",
                "Can you help?",
                "What can you do for me?",
                "I need support",
                "I need help",
                "Support me please"
            ],
            "responses": [
                "Tell me how I can assist you.",
                "What’s your problem? I’ll help you.",
                "Yes, sure! How can I support you?"
            ]
        }
    ]
}


import json
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
import pickle
# Save intents to JSON file
with open('intents.json', 'w') as outfile:
    json.dump(data, outfile, indent=4)

print("Intents saved to 'intents.json'.")

# Data Preparation
training_sentences = []
training_labels = []
labels = []
responses = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        training_sentences.append(pattern)
        training_labels.append(intent['tag'])
    responses.append(intent['responses'])

    if intent['tag'] not in labels:
        labels.append(intent['tag'])

num_classes = len(labels)

# Encode labels
lbl_encoder = LabelEncoder()
lbl_encoder.fit(training_labels)
training_labels_encoded = lbl_encoder.transform(training_labels)
```

```python
# Save the label encoder
with open('label_encoder.pickle', 'wb') as enc:
    pickle.dump(lbl_encoder, enc, protocol=pickle.HIGHEST_PROTOCOL)

print("LabelEncoder saved to 'label_encoder.pickle'.")

# Tokenization and Padding
vocab_size = 1000
embedding_dim = 16
max_len = 20
oov_token = "<OOV>"

tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(training_sentences)
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len)
```

```python
# Save the tokenizer
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)

print("Tokenizer saved to 'tokenizer.pickle'.")

# Model Building
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam', metrics=['accuracy'])

# Model Summary
model.summary()

# Training the Model
epochs = 500  # Consider reducing this number to prevent overfitting
history = model.fit(padded_sequences, np.array(training_labels_encoded), epochs=epochs)

# Save the trained model
model.save("chat_model.h5")
print("Model saved to 'chat_model.h5'.")
```

```python
import json
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder
import random
import pickle
# Load intents
with open("intents.json") as file:
    data = json.load(file)

# Load the trained model
model = keras.models.load_model('chat_model.h5')
print("Model loaded.")

# Load the tokenizer
with open('tokenizer.pickle', 'rb') as handle:
    tokenizer = pickle.load(handle)
print("Tokenizer loaded.")

# Load the label encoder
with open('label_encoder.pickle', 'rb') as enc:
    lbl_encoder = pickle.load(enc)
print("LabelEncoder loaded.")

def chat():
    max_len = 20  # Must be same as used during training

    while True:
        print("User: ", end="")
        inp = input()
        if inp.lower() == "quit":
            print("ChatBot: Goodbye! Have a great day!")
            break

        # Preprocess the input
        sequences = tokenizer.texts_to_sequences([inp])
        padded = keras.preprocessing.sequence.pad_sequences(sequences, truncating='post', maxlen=max_len)

        # Predict the intent
        result = model.predict(padded)
        predicted_label = lbl_encoder.inverse_transform([np.argmax(result)])[0]

        # Find the corresponding intent
        for intent in data['intents']:
            if intent['tag'] == predicted_label:
                response = random.choice(intent['responses'])
                print("ChatBot:", response)
                break
        else:
            # Fallback response if no intent is matched
            print("ChatBot: I'm sorry, I didn't understand that. Could you please rephrase?")

print("Start messaging with the bot (type 'quit' to stop)!")
chat()
```
